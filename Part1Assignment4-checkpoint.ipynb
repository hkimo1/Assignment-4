```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
class AdvData:
      def __init__(self, file_path):
        # Read the data from the CSV file
        self.t = np.array(file_path['time'])  # 'time' is the column name
        self.u = np.array(file_path['U'])      # 'U' is the magnitude column name
        self.u_filtered = None            
      
      def plot_time_series(self):
          plt.figure(figsize=(8, 4))
          plt.plot(self.t, self.u, label='Velocity')
          plt.title('Time Series Plot of Velocity')
          plt.xlabel('Time(s)')
          plt.ylabel('Velocity(m/s)')
          plt.legend()
          plt.show()
         
      
      def plot_histogram(self):
          
          plt.figure(figsize=(8, 4))
          plt.hist(self.u,  label='Velocity' )  
          plt.title('Time vs Velocity histogram')
          plt.xlabel('Velocity(m/s)')
          plt.ylabel('Frequency')
          plt.legend()
          plt.show()

      def apply_threshold(self, threshold):
            if self.u is not None:
                self.u_filtered = np.copy(self.u)
        # Apply the threshold condition and set "bad" data points to np.nan
                self.u_filtered[self.u > threshold] = np.nan
        
        # Calculate the percentage of NaN values
                num_nan = np.sum(np.isnan(self.u_filtered))
                percentage_nan = (num_nan / len(self.u)) * 100

                return percentage_nan
            else:
                print("No data available.")
                return None


      def plot_filtered_signal(self):
            if self.u is not None and self.u_filtered is not None:
               plt.figure(figsize=(8, 5))
               plt.plot(self.t, self.u, label='Original Signal', color='blue', alpha=0.6)# Plot original signal
               plt.plot(self.t, self.u_filtered, label='Filtered Signal', color='red', alpha=0.8)
               plt.title("Comparison of Original and Filtered Signals")
               plt.xlabel("Time (s)")
               plt.ylabel("Velocity (m/s)")
               plt.legend()
               plt.grid(alpha=0.3)
               plt.show()
            else:
                print("Original or filtered data is missing.")
            return None
          
file_path = pd.read_csv(r"C:\Users\Hayat\Documents\adv.csv")
adv_data = AdvData(file_path)
threshold = 1
percentage_nan = adv_data.apply_threshold(threshold)

print(adv_data.plot_time_series())
print(adv_data.plot_histogram())
print(f"Percentage of data set to NaN: {percentage_nan:.2f}%")
print(adv_data.plot_filtered_signal())
```
